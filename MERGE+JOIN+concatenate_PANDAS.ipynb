# Merge

```python
left=pd.DataFrame({"id":[1,2,3,4,5],
                   "Name":["Pers1","Pers2","Pers3","Pers4","Pers5"],
                   "U_id":["UID1","UID2",'UID3',"UID4","UID5"]})
right=pd.DataFrame({"id":[1,2,3,4,5],
                   "Name":["Pers6","Pers7","Pers8","Pers9","Pers10"],
                   "U_id":["UID2","UID4",'UID3',"UID6","UID5"]})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how="inner",left_index=True,right_index=True)
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on="U_id")            # bydefault inner 
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on="U_id",how="left")
```

```python
pd.merge(left,right,on="U_id",how="right")
```

```python
pd.merge(left,right,on="U_id",how="outer")
```

# JOINS
# combine the dataframe based on the index

```python
left.join(right,lsuffix="first_df",rsuffix="second_df")
```
